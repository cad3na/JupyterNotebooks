# Comparación de RL contra Realización de Estado

## Root Locus

Tenemos el siguiente sistema retroalimentado:

$$
G(s) = \frac{4}{s(s+2)}
$$

```python
G = tf([0, 0, 1],[1, 2, 0])
```

```python
figure(figsize=(10, 10))
r, k = rlocus(G, linspace(0, 100, 1000))
ax = gca()
ax.set_xlim(-2.5, 0.5)
ax.set_ylim(-3.0, 3.0);
```

Y utilizando un compensador:

$$
G_C(s) = \frac{s + 2.9}{s + 5.4}
$$

```python
Gc = tf([0, 1, 2.9], [0, 1, 5.4])
K = tf([0, 0, 4], [0, 0, 1])
```

```python
figure(figsize=(10, 10))
r, k = rlocus(Gc*K*G, linspace(0, 100, 1000))
ax = gca()
ax.set_xlim(-6.5, 0.5)
ax.set_ylim(-4.0, 4.0);
```

Y graficando ambos juntos:

```python
figure(figsize=(10, 10))
r, k = rlocus(G, linspace(0, 1000, 10000))
r, k = rlocus(Gc*K*G, linspace(0, 1000, 10000))
ax = gca()
ax.set_xlim(-6.5, 0.5)
ax.set_ylim(-4.0, 4.0);
```

## Realización de Estado

Ahora, para analizar el mismo sistema con su representación de estado, primero lo tenemos que llevar a esta representación:

Primero notamos que el grado de numerador es $m = 0$ y que el grado del denominador es $n = 2$, por lo que, de acuerdo a nuestra forma normal para las Ecuaciones Diferenciales Ordinarias, tenemos que los coeficientes de la EDO son:

$$
\frac{y(s)}{u(s)} =
\frac{b_0 s^m + b_1 s^{m-1} + ... + b_{m-1} s + b_m}{s^n + a_1 s^{n-1} + ... + a_{n-1} + a_n}
$$

Si $G(s) = \frac{4}{s^2 + 2s}$, entonces:

$$
b_0 = 4
$$

$$
a_1 = 2 , \quad a_2 = 0
$$

Por lo que tenemos que nuestra representación de estado será la siguiente:

$$
\frac{d}{dt}
\begin{pmatrix}
    x_1 \\
    x_2
\end{pmatrix} = 
\begin{pmatrix}
    0 & 1 \\
    0 & -2
\end{pmatrix}
\begin{pmatrix}
    x_1 \\
    x_2
\end{pmatrix} +
\begin{pmatrix}
    0 \\
    1
\end{pmatrix} u
$$

$$
y =
\begin{pmatrix}
    1 & 0 
\end{pmatrix}
\begin{pmatrix}
    x_1 \\
    x_2
\end{pmatrix}
$$

```python
var("s")
```

```python
A = Matrix([[0, 1], [0, -2]])
b = Matrix([[0],[1]])
c = Matrix([[4], [0]])
d = 0
```

Podemos obtener su función de transferencia con la siguiente formula:

$$
F(s) = c^T (sI - A)^{-1} b
$$

```python
G = c.T*(s*eye(2) - A).inv()*b
```

```python
G[0]
```

Y obtenemos su polinomio caracteristico con:

$$
P(s) = |s I - A|
$$

y sus polos con los eigenvalores de $A$.

```python
Pc = (s*eye(2) - A).det()
Pc
```

```python
A.eigenvals()
```

Los ceros del sistema se pueden obtener con el determinante de la matriz sistema.

$$
\Sigma (s) =
\begin{pmatrix}
    sI - A & b \\
    -c^T & d
\end{pmatrix}
$$

```python
S =((s*eye(2) - A).row_join(b)).col_join((-c.T).row_join(Matrix([d])))
S
```
