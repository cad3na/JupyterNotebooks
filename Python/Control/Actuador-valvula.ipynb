```python
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
from control import tf, step, pade
```

```python
from numpy import linspace
```

```python
t = linspace(0, 5, 1000)
```

```python
Gact = tf([1],[0.052267, 0.669, 1])*tf(pade(0.1, 3)[0], pade(0.1, 3)[1])
Gact
```

```python
y, t = step(Gact, t)
fig = figure(figsize=(22, 5))
ax = fig.gca()
ax.plot(t, y)
```

```python
Gval = tf([1.25], [3, 1])
```

```python
y, t = step(Gval, t)
fig = figure(figsize=(22, 5))
ax = fig.gca()
ax.plot(t, y)
```

```python
y, t = step(10*Gval*Gact, t)
fig = figure(figsize=(22, 5))
ax = fig.gca()
ax.plot(t, y)
```
