```python
from control import tf, rlocus, bode, step_response
from matplotlib.pyplot import style, rcParams, figure
%matplotlib inline
style.use("ggplot")
rcParams['figure.figsize'] = [18, 5]
```

```python
G = tf([1], [1,2,-2,1])
t,y= step_response(G)
```

```python
fig = figure()
ax = fig.gca()
ax.plot(t,y)
```

```python
Kd, Kp = 8,10
Con  = tf([Kd, Kp],[1])
t,y = step_response((Con*G).feedback(tf([2],[1])))
```

```python
fig = figure()
ax = fig.gca()
ax.plot(t,y)
```
