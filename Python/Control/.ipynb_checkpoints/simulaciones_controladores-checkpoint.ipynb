```python
from scipy.integrate import ode
from numpy import cos, sin, pi, linspace, array
```

```python
%matplotlib inline
from matplotlib.pyplot import figure, style
style.use("ggplot")
```

```python
class pendulo_simple:
    def __init__(self, masa, longitud, inercia, tipo_control):
        self.g = 9.81
        self.m = masa
        self.l = longitud
        self.J = inercia
        
        if tipo_control is "nulo":
            self.control = self.control_nulo
            
        if tipo_control is "PD":
            self.control = self.control_PD
            self.kp = 10
            self.kd = 2
            self.qd = 1
            
        if tipo_control is "PD_tray":
            self.control = self.control_PD_tray
            self.kp = 10
            self.kd = 2
            self.ω = 2*pi
        
    def ec_movimiento(self, t, x):
        q, q̇ = x
        τ = self.control(t, x)
        return [q̇, (τ - self.m*self.g*self.l*cos(q))/(self.m*self.l**2 + self.J)]
    
    def trayectoria(self, t):
        return sin(self.ω*t)
    
    def log_estado(self, x):
        self.estado.append(x)
        
    def log_tiempo(self, t):
        self.tiempo.append(t)
    
    def control_nulo(self, t, x):
        return 0
    
    def control_PD(self, t, x):
        q, q̇ = x
        return self.kp*(self.qd - q) + self.kd*(0 - q̇)
    
    def control_PD_tray(self, t, x):
        q, q̇ = x
        qd = self.trayectoria(t)
        return self.kp*(qd - q) + self.kd*(0 - q̇)
```

```python
robot1 = pendulo_simple(0.5, 0.4, 0.002, "nulo")

t0, tf, x0 = 0, 10, [0, 0]
n = 1000
δt = (tf - t0)/n

sis1 = ode(robot1.ec_movimiento)
sis1.set_initial_value(x0, t0)

xs1 = [array(x0)]
ts1 = [t0]
for i in range(n):
    xs1.append(sis1.integrate(sis1.t + δt))
    ts1.append(sis1.t)
```

```python
robot2 = pendulo_simple(0.5, 0.4, 0.002, "PD")

t0, tf, x0 = 0, 10, [0, 0]
n = 1000
δt = (tf - t0)/n

sis2 = ode(robot2.ec_movimiento)
sis2.set_initial_value(x0, t0)

xs2 = [array(x0)]
ts2 = [t0]
for i in range(n):
    xs2.append(sis2.integrate(sis2.t + δt))
    ts2.append(sis2.t)
```

```python
fig = figure(figsize=(17,4))
ax = fig.gca()
ax.plot(ts1, array(xs1).T[0]);
```

```python
fig = figure(figsize=(17,4))
ax = fig.gca()
ax.plot(ts2, array(xs2).T[0]);
```

```python
robot3 = pendulo_simple(0.5, 0.4, 0.002, "PD_tray")

t0, tf, x0 = 0, 10, [0, 0]
n = 1000
δt = (tf - t0)/n

sis3 = ode(robot3.ec_movimiento)
sis3.set_initial_value(x0, t0)
sis3.set_integrator('vode', nsteps=100, method="bdf")

xs3 = [array(x0)]
ts3 = [t0]
for i in range(n):
    xs3.append(sis3.integrate(sis3.t + δt))
    ts3.append(sis3.t)
```

```python
us3 = [robot3.control(0, estado) for estado in xs3]
tray3 = [robot3.trayectoria(t) for t in ts3]
```

```python
fig = figure(figsize=(17,8))
ax1, ax2 = fig.subplots(2,1)
ax1.plot(ts3, array(xs3).T[0])
ax1.plot(ts3, tray3)
ax2.plot(ts3, us3);
```
