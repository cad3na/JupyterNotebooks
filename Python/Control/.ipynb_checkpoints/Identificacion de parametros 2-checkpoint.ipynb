```python
def escalon(x, t, mag):
    if t <= 0:
        y = 0
    else:
        y = mag
    return y

def senoidal(x, t, mag, frec):
    from numpy import sin
    if t <=  0:
        y = 0
    else:
        y = mag*sin(frec*t)
    return y

def motor_cd(x, t, entrada):
    y, ẏ = x
    a, b = 2, 50
    u = entrada(x, t)
    us.append([t,u])
    ÿ = b*u - a*ẏ
    return ẏ, ÿ
```

```python
ts = linspace(0, 10, 100)
x0 = [0, 0]
u1 = lambda x, t: escalon(x, t, 3)
us = []
datos = odeint(motor_cd, x0, ts, args=(u1,))
y, yp = zip(*datos)
```

```python
us.sort()
```

```python
tus, uus = zip(*us)
prueba = interp(ts, tus, uus)
prueba
```

```python
fig = figure(figsize=(18,6))
ax1, ax2, ax3 = fig.subplots(1,3)
p1, = ax1.plot(ts, y)
p2, = ax2.plot(ts, yp)
p3, = ax3.plot(ts, prueba)
ax1.legend([p1], [r"$q$"])
ax2.legend([p2], [r"$\dot{q}$"])
ax3.legend([p3], [r"$u$"]);
```

```python
def control_pd(x, t, referencia, kp = 1, kd = 0):
    y, ẏ = x
    yd = referencia(x, t)
    refs.append([t, yd])
    e = yd - y
    es.append([t, e])
    ė = 0 - ẏ
    salida = kp*e + kd*ė
    return salida
```

```python
ts = linspace(0, 10, 100)
x0 = [0, 0]
en1 = lambda x, t: escalon(x, t, 1)
u1 = lambda x, t: control_pd(x, t, en1, 9, 2)
us = []
es = []
refs = []
datos = odeint(motor_cd, x0, ts, args=(u1,))
y, yp = zip(*datos)
```

```python
us.sort()
es.sort()
refs.sort()
```

```python
tus, uus = zip(*us)
señ_control = interp(ts, tus, uus)

tes, ees = zip(*es)
errores = interp(ts, tes, ees)

trefs,rrefs = zip(*refs)
referencia = interp(ts, trefs, rrefs)
```

```python
fig = figure(figsize=(18,8))
[ax1, ax2], [ax3, ax4] = fig.subplots(2,2)
p1, = ax1.plot(ts, y)
pr, = ax1.plot(ts, referencia)
p2, = ax2.plot(ts, yp)
p3, = ax3.plot(ts, señ_control)
p4, = ax4.plot(ts, errores)
ax1.legend([p1, pr], [r"$q$", r"$\hat{q}$"])
ax2.legend([p2], [r"$\dot{q}$"])
ax3.legend([p3], [r"$u$"])
ax4.legend([p4], [r"$e = q - \hat{q}$"]);
```

```python
def filtro_primer_orden(estado, t, entrada, a):
    y = estado
    u = entrada(estado, t)
    salida = a*u - a*y
    return salida
```

```python
from numpy.random import normal
from numpy import linspace
from scipy.signal import resample
```

```python
t0, tf = 0, 10
periodo_integracion = 0.01

muestras = int(1/periodo_integracion)

ruido = normal(0, 0.001, muestras)
ts = linspace(t0, tf, muestras)

ruido, ts = resample(ruido, muestras*10+1, ts)
```

```python
len(ruido)
```

```python
fig = figure(figsize=(18,8))
ax1 = fig.subplots(1,1)
p1, = ax1.plot(ts, ruido)
ax1.legend([p1], [r"$\hat{q}$"]);
ax1.set_xlim(t0, tf);
```

```python
def ruido(t0, tf, h):
    from numpy.random import normal
    from numpy import linspace
    from scipy.signal import resample
    
    t = h*10
    n = int(1/t)
    ruido = normal(0, 0.01, n)
    ts = linspace(t0, tf, n)

    ruido, ts = resample(ruido, n*10+1, ts)
    
    return ts, ruido
```

```python
ts, en1 = ruido(0, 10, 0.001)
fig = figure(figsize=(18,8))
ax1 = fig.subplots(1,1)
p1, = ax1.plot(ts, en1)
ax1.legend([p1], [r"$\hat{q}$"]);
ax1.set_xlim(t0, tf);
```

```python
def alimentador(x, t, ts, xs):
    from numpy import interp
    salida = interp(t, ts, xs)
    return salida
```

```python
ts, ss = ruido(0, 10, 0.001)
x0 = [0, 0]
s1 = lambda x, t: alimentador(x, t, ts, ss)
datos = odeint(filtro_primer_orden, x0, ts, args=(s1, 20))
s2, _ = zip(*datos)
```

```python
fig = figure(figsize=(18,8))
ax1 = fig.subplots(1,1)
p1, = ax1.plot(ts, ss)
p2, = ax1.plot(ts, s2)
ax1.legend([p1, p2], [r"$\hat{q}$", r"$\hat{q}$"]);
ax1.set_xlim(t0, tf);
```
