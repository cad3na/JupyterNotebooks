```python
from numpy import matrix, pi, linspace
τ = 2*pi
```

```python
t0, tf = 0, 10
M = matrix([[1, t0, t0**2, t0**3],
            [0, 1, 2*t0, 3*t0**2],
            [1, tf, tf**2, tf**3],
            [0, 1, 2*tf, 3*tf**2]])
M
```

```python
q0, qf = 0, τ/4
v0, vf = 0, 0
b = matrix([[q0], [v0], [qf], [vf]])
b
```

```python
a = M.I*b
coefs = a.T.tolist()[0]
coefs
```

```python
poli_3 = lambda coef, x: coef[0] + coef[1]*x + coef[2]*x**2 + coef[3]*x**3
q = lambda t: poli_3(coefs, t)
q̇ = lambda t: poli_3([coefs[1], 2*coefs[2], 3*coefs[3], 0], t)
q̈ = lambda t: poli_3([2*coefs[2], 6*coefs[3], 0, 0], t)
```

```python
t = linspace(0, 12, 500)
y = q(t)
ẏ = q̇(t)
ÿ = q̈(t)
```

```python
from matplotlib.pyplot import figure, style
%matplotlib notebook
style.use("ggplot")
```

```python
fig = figure(figsize=(8,4))
ax = fig.gca()

ax.plot(t, y)
ax.plot(t, ẏ)
ax.plot(t, ÿ)
```
