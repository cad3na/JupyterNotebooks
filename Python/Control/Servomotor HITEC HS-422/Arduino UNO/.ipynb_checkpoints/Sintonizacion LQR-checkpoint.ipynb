```python
a, b = 12, 8
A = matrix([[0, 1], [0, -a]])
B = matrix([[0], [b]])
C = matrix([[1, 0]])
D = matrix([[0]])
```

```python
servo_HS = ss(A, B, C, D)
```

```python
Q, R = matrix([[10, 0], [0, 1]]), matrix([[1]])
K, S, E = lqr(servo_HS, Q, R)
```

```python
def escalon(t, mag):
    '''
    Esta función toma un vector con el estado del sistema, el tiempo de simulación
    y la magnitud del escalon a simular, y da como resultado el valor de la función
    escalón en el tiempo requerido.
    '''
    if t <= 0:
        y = 0
    else:
        y = mag
    return y

def motor_cd_simple(t, x, control, a, b):
    '''
    Esta función toma un vector con el estado del sistema, el tiempo de simulación, 
    y nombre de función de entrada, y da como resultado la derivada del sistema,
    calculada a partir de la dinámica simplificada de un motor de corriente directa y de
    la señal de entrada del motor.
    '''
    y, ẏ = x
    u = control(t, x)
    ÿ = b*u - a*ẏ
    return [ẏ, ÿ]

def control_pd(t, x, entrada, kp, kd):
    p_sat = 255
    e = entrada(t) - x[0]
    ep = - x[1]
    u = kp*e - kd*ep
    return int(max(min(p_sat, u), -p_sat))
```

```python
s1 = lambda t: escalon(t, 127)
control_HS422 = lambda t, x: control_pd(t, x, s1, K[0][0], K[0][1])
motor_HS422 = lambda t, x: motor_cd_simple(t, x, control_HS422, a, b)
```

```python
sis = ode(motor_HS422)
sis.set_initial_value([0, 0])
sis.set_integrator('vode', nsteps=100, method="bdf");
```

```python
t_sim = linspace(0, 10, 1000)
y_sim = [sis.integrate(t) for t in t_sim[1:]]
y_sim.insert(0, [0, 0])
```

```python
fig = figure(figsize=(16, 4))
ax = fig.gca()
ax.plot(t_sim, array(y_sim).T[0]);
```

```python
def alg_control(xs, t):
    kp, kd = 10, 2
    p_sat = 255
    e = referencias(t) - xs[-1]
    ep = xs[-1] - xs[-2]
    u = kp*e - kd*ep
    return int(max(min(p_sat,u),-p_sat))

def referencias(t):
    from numpy import interp
    if t >= 5:
        return 127
    else:
        return 0
```

```python
from serial import Serial
from time import time, sleep
```

```python
xs = []
us = []
ts = []

with Serial("/dev/cu.usbserial-A603QGMH", baudrate=115200) as puerto_arduino:
    t0 = time()
    while time() - t0 < 15:
        puerto_arduino.reset_input_buffer()
        lec = int.from_bytes(puerto_arduino.read(1), byteorder='little')
        xs.append(lec)
        ts.append(time() - t0)
        
        if len(xs) > 1:
            us.append(alg_control(xs[-2:], ts[-1]))
        else:
            us.append(0)
            
        if us[-1] > 0:
            puerto_arduino.write(bytes([0]))
            puerto_arduino.write(bytes([us[-1]]))
        else:
            puerto_arduino.write(bytes([1]))
            puerto_arduino.write(bytes([abs(us[-1])]))
```

```python
fig = figure(figsize=(16, 4))
ax = fig.gca()
ax.plot(ts, array(xs));
```

```python
t_exp = []
q_exp = []
u_exp = []
for i, t in enumerate(ts):
    if t > 5:
        t_exp.append(ts[i])
        q_exp.append(xs[i])
        u_exp.append(us[i])
```

```python
t_interp = linspace(5, 15, 1000)
q_interp = array([interp(t, t_exp, q_exp) for t in t_interp])
u_interp = array([interp(t, t_exp, u_exp) for t in t_interp])
```

```python
fig = figure(figsize=(16, 4))
ax = fig.gca()
ax.plot(t_interp, q_interp);
```
