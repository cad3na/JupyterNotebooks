```python
def generacion_trayectorias_senoidal(t0, t1, n, A, ω, μ):
    from numpy import sin, linspace
    from control import forced_response, tf
    
    f = lambda t: A*sin(ω*t) + μ
    ts = linspace(t0, t1, n)
    us = [f(t) for t in ts]
    
    fso0 = tf([400], [1, 40, 400])
    fso1 = tf([400, 0], [1, 40, 400])
    fso2 = tf([400, 0, 0], [1, 40, 400])
    
    ts, nus, xs = forced_response(fso0, ts, us)
    ts, nu̇s, xs = forced_response(fso1, ts, us)
    ts, nüs, xs = forced_response(fso2, ts, us)
    
    return ts, nus, nu̇s, nüs

class motor_cd:
    def __init__(self, a, b, kp, kd, tipo_control, tiempos, trayectorias):
        from numpy import sin, cos, pi, interp
        self.a = a
        self.b = b
        self.kp = kp
        self.kd = kd
        self.ts = tiempos
        self.p_sat = 255
        self.qds, self.q̇ds, self.q̈ds = trayectorias
            
        if tipo_control is "PD_mas":
            self.control = self.control_PD_mas
            #self.λ = self.kp/self.kd
            self.λ = 0
        
    def ec_movimiento(self, t, x):
        q, q̇ = x
        τ = self.control(t, x)
        return [q̇, self.b*τ - self.a*q̇]
    
    def inter_tray(self, t):
        from numpy import interp
        return interp(t, self.ts, self.qds), interp(t, self.ts, self.q̇ds), interp(t, self.ts, self.q̈ds)
    
    def control_PD_mas(self, t, x):
        q, q̇ = x
        qd, q̇d, q̈d = self.inter_tray(t)
        m = 1/self.b
        c = self.a/self.b
        g = 0
        e, ė = qd - q, q̇d - q̇
        u = m*q̈d + c*q̇d + g + (self.kd + m*self.λ)*ė + (self.kp + c*self.λ)*e
        return int(max(min(self.p_sat, u), -self.p_sat))
```

```python
t0, tf, x0 = 0, 10, [0, 0]
n = 1000
δt = (tf - t0)/n

t_tray, tray_q, tray_q̇, tray_q̈ = generacion_trayectorias_senoidal(t0, tf, n, 50, 1*pi, 127)

fig = figure(figsize=(20,4))
ax1 = fig.subplots(1,1)
p2, = ax1.plot(t_tray, tray_q)
```

```python
a, b = matrix([[14.27545351],[27.54070936]])
kp, kd = 10, 2.79751187
```

```python
servo_HS422_PD_mas = motor_cd(a, b, kp, kd, "PD_mas", t_tray, [tray_q, tray_q̇, tray_q̈])

sis = ode(servo_HS422_PD_mas.ec_movimiento)
sis.set_initial_value(x0, t0)
sis.set_integrator('vode', nsteps=500, method="bdf")

t_sim = [t0]
x_sim = [array(x0)]

while t_sim[-1] < tf:
    x_sim.append(sis.integrate(t_sim[-1] + δt))
    t_sim.append(sis.t)
```

```python
x_sim = array(x_sim)
t_sim = array(t_sim)
```

```python
u_sim = [servo_HS422_PD_mas.control(t, x) for t, x in zip(t_sim, x_sim)]
e_sim = array([interp(t, t_tray, tray_q) for t in t_sim]) - x_sim.T[0]
```

```python
fig = figure(figsize=(20,10))
ax1, ax2, ax3 = fig.subplots(3,1)
p1, = ax1.plot(t_sim, x_sim.T[0])
p2, = ax1.plot(t_tray, tray_q)
p3, = ax2.plot(t_sim, u_sim)
p4, = ax3.plot(t_sim, e_sim)
ax1.legend([p1, p2],["posicion", "trayectoria deseada"])
ax2.legend([p3], ["señal de control"])
ax3.legend([p4], ["error"]);
```

```python
def inter_tray(t, ts, tras, traps, trapps):
    from numpy import interp
    if t > 3:
        return interp(t-3, ts, tras), interp(t-3, ts, traps), interp(t-3, ts, trapps)
    else:
        return 0,0,0
    
def alg_control(xs, ts, kp, kd):
    p_sat = 255
    qd, q̇d, q̈d = inter_tray(ts[-1], t_tray, tray_q, tray_q̇, tray_q̈)
    e = qd - xs[-1]
    ė = q̇d - (xs[-1] - xs[-2])
    m = 1/b
    c = a/b
    g = 0
    u = m*q̈d + c*q̇d + g + (kd + m*0)*ė + (kp + c*0)*e
    return int(max(min(p_sat,u),-p_sat))
```

```python
from serial import Serial
from time import time, sleep
```

```python
kp, kd = 15, 0.279

x_real = [0]
u_real = [0]
t_real = [0]

with Serial("/dev/cu.usbmodem0E204D821", baudrate=57600) as puerto_arduino:
    while t_real[-1] < 15:
        try:
            lec = int.from_bytes(puerto_arduino.read(6), byteorder='little')
            x_real.append(int(lec/(2**32))/16)
            t_real.append(int(lec%(2**32))*0.000001)
        except KeyboardInterrupt:
            break
        
        if len(x_real) > 1:
            u_real.append(alg_control(x_real[-2:], t_real[-2:], kp, kd))
        else:
            u_real.append(0)
            
        if u_real[-1] > 0:
            puerto_arduino.write(bytes([0, u_real[-1]]))
        else:
            puerto_arduino.write(bytes([1, abs(u_real[-1])]))
            
    puerto_arduino.write(bytes([0, 0]))
```

```python
t_exp = []
q_exp = []
u_exp = []
for i, t in enumerate(t_real):
    if t > 3:
        t_exp.append(t_real[i])
        q_exp.append(x_real[i])
        u_exp.append(u_real[i])
```

```python
t_interp = linspace(3, 13, 1000)
q_interp = array([interp(t, t_exp, q_exp) for t in t_interp])
u_interp = array([interp(t, t_exp, u_exp) for t in t_interp])
```
