```python
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
from control import tf, step, pade
```

```python
from numpy import linspace
```

```python
t = linspace(0, 5, 1000)
```

```python
Gboi = tf([1],[1, 1/60])*tf(pade(1, 20)[0], pade(1, 20)[1])
Gboi
```

```python
y, t = step(Gboi, t)
fig = figure(figsize=(22, 5))
ax = fig.gca()
ax.plot(t, y)
```

```python
Gval = tf([1.25], [0.1, 1])
```

```python
y, t = step(10*Gval*Gboi, t)
fig = figure(figsize=(22, 5))
ax = fig.gca()
ax.plot(t, y)
```
