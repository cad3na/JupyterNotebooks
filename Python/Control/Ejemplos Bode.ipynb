```python
from control import tf, rlocus, bode
from numpy import logspace
from matplotlib.pyplot import style, rcParams
%matplotlib inline
#style.use("ggplot")
rcParams['figure.figsize'] = [9, 6]
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = 'Noto Sans'
rcParams['font.size'] = 16.0
```

```python
G1 = tf([10],[1])
mag, ϕ, ω = bode(G1, dB=True)
```

```python
G2 = tf([1,0],[1])
mag, ϕ, ω = bode(G2, dB=True, color="k")
```

```python
G3 = tf([1],[1,0])
mag, ϕ, ω = bode(G3, dB=True, color="k")
```

```python
G4 = tf([1,1],[1])
mag, ϕ, ω = bode(G4, dB=True)
```

```python
G5 = tf([1],[1,1])
mag, ϕ, ω = bode(G5, dB=True, color="k")
```

```python
G6 = tf([1],[1,0])*tf([1],[1,10,1])
ω = logspace(-1, 1, 1000)
mag, ϕ, ω = bode(G6, omega=ω, dB=True, color="k")
```
