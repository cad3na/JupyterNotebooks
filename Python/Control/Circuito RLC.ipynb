```python
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
from control import tf, step, forced_response, impulse
```

```python
from numpy import linspace, sin, sqrt
```

```python
C = 0.000000001
R = 100
L = 0.0010
```

```python
ωₙ = 1/(sqrt(L*C))
T = 1/ωₙ
ωₙ, T
```

```python
Grlc = tf([1/L, 0], [1, R/L, 1/(L*C)])
Grlc
```

```python
t = linspace(0, 20*T, 1000)
entrada = sin(t)
```

```python
#y, t = step(Grlc, t)
t, y, x = forced_response(Grlc, t, entrada, 0)
#y, t = impulse(Grlc, t)
```

```python
fig = figure(figsize=(22,5))
ax = fig.gca()
ax.plot(t, y)
#ax.plot(t, entrada)
```
