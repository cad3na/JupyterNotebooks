```python
from control import tf, forced_response
```

```python
G1 = tf([50], [1, 0.2])
G2 = G1.feedback(0.2)
```

```python
G3 = G2*tf([1], [1,0])
```

```python
G4 = G3.feedback(1)
```

```python
from numpy import sin, linspace
```

```python
ts = linspace(0, 10, 1000)
es = sin(ts)
```

```python
t, y, x = forced_response(G4, ts, es)
```

```python
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
fig = figure(figsize=(16,8))
ax = fig.gca()
ax.plot(ts, es)
ax.plot(t, y)
```
