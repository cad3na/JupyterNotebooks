# Balanceo de Ecuaciones

Tenemos la siguiente formula quimica que balancear:

$$
x_1 B_2 S_3 + x_2 H_2 O \implies x_3 H_3 B O_3 + x_4 H_2 S
$$

donde $x_1$, $x_2$, $x_3$ y $x_4$ son incognitas. Esto lo podemos describir con vectores de la siguiente manera:

$$
x_1
\begin{pmatrix}
    2 \\
    3 \\
    0 \\
    0
\end{pmatrix}+
x_2
\begin{pmatrix}
    0 \\
    0 \\
    2 \\
    1
\end{pmatrix}+
x_3
\begin{pmatrix}
    1 \\
    0 \\
    3 \\
    3
\end{pmatrix}+
x_4
\begin{pmatrix}
    0 \\
    1 \\
    2 \\
    0
\end{pmatrix}=
\begin{pmatrix}
    0 \\
    0 \\
    0 \\
    0
\end{pmatrix}
$$

Lo cual se puede ver como el siguiente sistema lineal:

$$
\begin{pmatrix}
2 & 0 & 1 & 0 \\
3 & 0 & 0 & 1 \\
0 & 2 & 3 & 2 \\
0 & 1 & 3 & 0
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2 \\
x_3 \\
x_4
\end{pmatrix} =
\begin{pmatrix}
0 \\
0 \\
0 \\
0
\end{pmatrix}
$$

Podemos resolver este sistema de ecuaciones; como es homogeneo, lo podemos representar con una matriz cuadrada y tan solo tenemos que hacer una simplificación Gaussiana:

$$
\begin{pmatrix}
2 & 0 & 1 & 0 \\
3 & 0 & 0 & 1 \\
0 & 2 & 3 & 2 \\
0 & 1 & 3 & 0
\end{pmatrix}
\implies
\begin{pmatrix}
2 & 0 & 1 & 0 \\
0 & 1 & 3 & 0 \\
0 & 2 & 3 & 2 \\
3 & 0 & 0 & 1
\end{pmatrix}
\implies
\begin{pmatrix}
1 & 0 & \frac{1}{2} & 0 \\
0 & 1 & 3 & 0 \\
0 & 2 & 3 & 2 \\
1 & 0 & 0 & \frac{1}{3}
\end{pmatrix}
\implies
\begin{pmatrix}
1 & 0 & \frac{1}{2} & 0 \\
0 & 1 & 3 & 0 \\
0 & 0 & -3 & 2 \\
0 & 0 & -\frac{1}{2} & \frac{1}{3}
\end{pmatrix}
\implies
\begin{pmatrix}
1 & 0 & \frac{1}{2} & 0 \\
0 & 1 & 3 & 0 \\
0 & 0 & 1 & -\frac{2}{3} \\
0 & 0 & -\frac{1}{2} & \frac{1}{3}
\end{pmatrix}
\implies
$$

$$
\begin{pmatrix}
1 & 0 & \frac{1}{2} & 0 \\
0 & 1 & 3 & 0 \\
0 & 0 & 1 & -\frac{2}{3} \\
0 & 0 & 0 & 0
\end{pmatrix}
\implies
\begin{pmatrix}
1 & 0 & \frac{1}{2} & 0 \\
0 & 1 & 0 & 2 \\
0 & 0 & 1 & -\frac{2}{3} \\
0 & 0 & 0 & 0
\end{pmatrix}
\implies
\begin{pmatrix}
1 & 0 & 0 & \frac{1}{3} \\
0 & 1 & 0 & 2 \\
0 & 0 & 1 & -\frac{2}{3} \\
0 & 0 & 0 & 0
\end{pmatrix}
$$

de donde obtenemos las siguientes relaciones:

$$
\begin{array}{r c l}
x_1 + \frac{1}{3} x_4 & = & 0 \\
x_2 + 2 x_4 & = & 0 \\
x_3 - \frac{2}{3} x_4 & = & 0 \\
0 & = & 0
\end{array}
$$

y si $x_4 = 3$, tendremos que:

$$
\begin{array}{r c l}
x_1 & = & -1 \\
x_2 & = & -6 \\
x_3 & = & 2 \\
0 & = & 0
\end{array}
$$
