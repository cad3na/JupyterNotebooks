```python
from numpy import linspace, tan, log, sin, sqrt, cos
from scipy.optimize import fsolve
from matplotlib.pyplot import figure
%matplotlib inline
sec = lambda x: 1/cos(x)
```

```python
a = lambda x: 1/(x+1)**2 - x
b = lambda x: sqrt((6 - x - x**3)/(4)) - x
c = lambda x: sin(x) - x
d = lambda x: log(x) - tan(x)
e = lambda x: 4 + (x - 1)/(x + 1) - x
f = lambda x: sec(x)/2 - x
```

```python
x = linspace(-10, 10, 100)
f1 = figure(figsize=(12,6))
a1 = f1.gca()
a1.plot(x, a(x))
a1.set_xlim(-10, 10)
a1.set_ylim(-1, 9);
fsolve(a, 0)
```

```python
x = linspace(-10, 10, 100)
f2 = figure(figsize=(12,6))
a2 = f2.gca()
a2.plot(x, b(x))
a2.set_xlim(-10, 10)
a2.set_ylim(-1, 9)
fsolve(b, 0)
```

```python
x = linspace(-10, 10, 100)
f3 = figure(figsize=(12,6))
a3 = f3.gca()
a3.plot(x, c(x))
a3.set_xlim(-10, 10)
a3.set_ylim(-1, 9)
fsolve(c, 0)
```

```python
x = linspace(-10, 10, 100)
f4 = figure(figsize=(12,6))
a4 = f4.gca()
a4.plot(x, d(x))
a4.set_xlim(-10, 10)
a4.set_ylim(-1, 9)
fsolve(d, 2)
```

```python
x = linspace(-10, 10, 100)
f5 = figure(figsize=(12,6))
a5 = f5.gca()
a5.plot(x, e(x))
a5.set_xlim(-10, 10)
a5.set_ylim(-1, 9);
fsolve(e, 0)
```

```python
x = linspace(-10, 10, 100)
f6 = figure(figsize=(12,6))
a6 = f6.gca()
a6.plot(x, f(x))
a6.set_xlim(-10, 10)
a6.set_ylim(-1, 9);
fsolve(f, 0)
```

```python
fsolve(a, 0), fsolve(b, 0), fsolve(c, 0), fsolve(d, 2), fsolve(e, 0), fsolve(f, 0)
```
