```python
A = matrix([[8,0,6,1,0,0,0,0,0,0,0],
            [0,9,0,0,0,0,5,0,2,1,0],
            [0,1,7,0,0,1,2,0,0,1,0],
            [0,1,0,6,0,0,1,0,0,0,1],
            [0,0,0,0,9,0,0,0,1,1,1],
            [0,1,0,2,0,10,1,0,3,0,0],
            [0,0,5,0,2,2,10,0,0,0,0],
            [0,0,0,6,1,0,0,15,0,2,0],
            [0,2,0,0,4,0,1,1,20,1,0],
            [0,0,0,0,0,3,0,6,5,25,1],
            [1,0,3,1,5,0,7,0,0,1,12]])
b = matrix([[1],[5],[8],[0],[8],[1],[0],[3],[0],[1],[2]])
```

```python
def jacobi(A, b, error):
    from numpy import matrix, zeros, copy, fill_diagonal, divide, sqrt
    x0 = matrix(zeros(A.shape[0])).T
    divisores = copy(A.diagonal().T)
    fill_diagonal(A, 0)
    xs = [x0]
    errores = [0]
    # Inicialize valores a utilizar y arreglos a llenar, como el anterior
    while True:
        # Escribe tu codigo aqui
        xs.append(divide(b, divisores) - divide(A*xs[-1], divisores))
        dif = xs[-1] - xs[-2]
        errores.append(sqrt((dif.T*dif)[0,0]))
        if errores[-1] < error:
            return xs, errores
```

```python
aproxs, errores = jacobi(A, b, 0.01)
```

```python
from numpy import hstack
```

```python
nombres = ["x01", "x02", "x03", "x04", "x05", "x06", "x07", "x08", "x09", "x10", "x11"]
resultados = DataFrame(hstack(aproxs).T, columns=nombres)
```
