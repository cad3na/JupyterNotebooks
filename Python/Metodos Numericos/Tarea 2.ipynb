```python
from numpy import exp, linspace, cos, log
from scipy.optimize import fsolve
from matplotlib.pyplot import figure
%matplotlib inline
```

$$
x e^x - 2 = 0
$$

```python
f = lambda x: x*exp(x) - 2
```

```python
r = fsolve(f, 0).tolist()[0]
r
```

```python
minx = -10
maxx = 1
steps = 100
x = linspace(minx, maxx, steps)
y = f(x)
miny = min(y)
maxy = max(y)
```

```python
fig = figure(figsize=(16, 6))
axis = fig.gca()
axis.plot([minx, maxx], [0, 0], "--")
axis.plot([r, r], [miny, maxy], "--")
axis.plot(x, y)
axis.set_xlim([minx, maxx])
axis.set_ylim([miny, maxy])
print("Raiz en:", "{:.3f}".format(r))
```
