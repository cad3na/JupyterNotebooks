Estime las siguientes integrales:

1. $$ \int_0^1 \sin{(101 \pi x)}dx $$
2. $$ \int_0^1 f(x)dx $$
con $ f(x) = \frac{\sin{x}}{x} $ en $ x \le 0 $ y $f(x) = 1$ en $ x=0 $.

```python
from numpy import sin, pi
g = lambda x: sin(101*pi*x)

def f(x):
    if x==0:
        return 1
    else:
        return sin(x)/x
```

```python
x0, xf = 0, 1
n = 11
```

```python
from numpy import linspace
```

```python
xs = linspace(x0, xf, n)
xs
```

```python
ys = [f(x) for x in xs]
ys[1:-1]
```

```python
def integral_trapecio(func, x0, xf, n):
    xs = linspace(x0, xf, n+1)
    ys = [func(x) for x in xs]
    h  = (xf - x0)/n
    I = h/2*(ys[0] + 2*sum(ys[1:-1]) + ys[-1] )
    return I
```

```python
integral_trapecio(f, 0, 1, 21)
```

```python
integral_trapecio(g, 0, 1, 20)
```
