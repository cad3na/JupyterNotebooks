```python
from matplotlib.pyplot import figure

from matplotlib.sankey import Sankey
```

```python
fig = figure(figsize=(11,8.5))
ax = fig.gca()

Sankey(ax=ax, flows=[0.25, 0.15, 0.60, -0.20, -0.15, -0.05, -0.50, -0.10],
       labels=['', '', '', 'First', 'Second', 'Third', 'Fourth', 'Fifth'],
       orientations=[-1, 1, 0, 1, 1, 1, 0, -1]).finish()

```

```python
fig = figure(figsize=(11,8.5))
ax = fig.add_subplot(1, 1, 1, xticks=[], yticks=[],
                     title="Flow Diagram of a Widget")

sankey = Sankey(ax=ax, scale=0.01, offset=0.2, head_angle=150,
                format='%.0f', unit='%')

sankey.add(flows=[25, 0, 60, -10, -20, -5, -15, -10, -40],
           labels=['', '', '', 'First', 'Second', 'Third', 'Fourth',
                   'Fifth', 'Hurray!'],
           orientations=[-1, 1, 0, 1, 1, 1, -1, -1, 0],
           pathlengths=[0.25, 0.25, 0.25, 0.25, 0.25, 0.6, 0.25, 0.25,
                        0.25],
           patchlabel="Widget\nA")  # Arguments to matplotlib.patches.PathPatch

diagrams = sankey.finish()

diagrams[0].texts[-1].set_color('r')
diagrams[0].text.set_fontweight('bold')
```

```python
fig = figure(figsize=(11, 8.5))#, dpi=300)
ax = fig.add_subplot(1, 1, 1, xticks=[], yticks=[])

sankey = Sankey(ax=ax, unit=None, head_angle=150, shoulder=0)

flows_1 = [0.25, 0.15, 0.60, -0.10, -0.05, -0.25, -0.15, -0.10, -0.35]
orientations_1 = [-1, 0, 0, 1, 1, 1, -1, -1, 0]
sankey.add(flows=flows_1, label='one',
           orientations=orientations_1)

flows_2 = [-0.25, 0.15, 0.1]
orientations_2 = [-1, -1, -1]
sankey.add(flows=flows_2, label='two',
           orientations=orientations_2, prior=0, connect=(0, 0))

diagrams = sankey.finish()
diagrams[-1].patch.set_hatch('/')
#ax.legend();
```
