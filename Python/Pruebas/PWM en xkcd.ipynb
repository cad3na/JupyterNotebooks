```python
%matplotlib inline

from matplotlib.pyplot import plot, figure, xkcd, ylim, ylabel, xlabel, xlim, grid
from numpy import linspace, pi
from scipy import signal
```

```python
xkcd()
figure(figsize=(10, 6))
t = linspace(-0.001, 2, 500, endpoint=False)
plot(t, 2.5*signal.square(3*pi*t)+2.5)
grid(True)
xlim(-0.1, 1.2)
ylim(-1, 6)
ylabel("Voltaje")
xlabel("Tiempo")
```
