```python
from tkinter import *
from tkinter import messagebox
from Crypto.Hash import SHA
from requests import Session
from bs4 import BeautifulSoup as bs
from re import findall
import datetime

def chck_temp():
    user="mexsat"
    pswd="mexsat"

    h=SHA.new()
    h.update(pswd.encode())

    pswd_sha=h.hexdigest()

    base_url="http://192.168.169.3:81/"
    login_data = {"uid_in":"mexsat","pid_in":"", "hex_in":pswd_sha}

    with Session() as s:
    
        response_header=login_data
    
        response=s.post(base_url + ".html?get=${|mexsat|}", login_data)
        doc=bs(response.text, 'html.parser')
        script=doc.find_all("script")[0].text
        pag, user, sessionID = findall(".*\$(\w*)\|(\w*)\|(\w*)';", script)[0]
        response_header = f".html?get=${pag}|{user}|{sessionID}"
        response=s.get(base_url + response_header)
        doc=bs(response.text, 'html.parser')
        script=doc.find_all("script")[0].text
        sessionID=findall('.*var sessionID = "([^\n\r]*)";', script)[0]
    
        param=[[f".html?get=$AHU01","0"],
               [f".html?get=$AHU02","0"],
               [f".html?get=$AHU03","0"],
               [f".html?get=$AHU04","0"],
               [f".html?get=$AHU05","0"],
               [f".html?get=$AHU10","2"],
               [f".html?get=$AHU11","37"],]
    
    
        Temp=[]
    
        for i in param:    
            response_header = i[0]+f"|{user}|{sessionID}"
        
            response=s.get(base_url + response_header)
            doc=bs(response.text, 'html.parser')
            script=doc.find_all("script")[0].text
            sessionID=findall('.*var sessionID = "([^\n\r]*)";', script)[0]
        
            cont=doc.find_all("body")[0]
            t=findall("id=\"tag"+i[1]+"\" nowrap=\"\" style=\"color: #0040e0\" width=\"100%\">[\n\r]\s*(\d{2})",cont.prettify())[0]
            ''.join(t)
            Temp.append(int(t))
    
        print(Temp)
    
        global RF1, RF2, RF3, RF4, RF5, SRVR1, SRVR2, actual
        RF1,RF2, RF3, RF4, RF5, SRVR1, SRVR2 = Temp
        
        actual = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        
        print(actual)

        
def insert():
    
    ult_act.set(actual)
    t_RF1.set(RF1)
    if RF1 > 22:
        Temp_value_RF1.config(bg = 'red')
        messagebox.showwarning("Warning","Temperatura RF1 elevada")
        
    else:
        Temp_value_RF1.config(bg = "#00FF00")
        
    t_RF2.set(RF2)
    if RF2 > 22:
        Temp_value_RF2.config(bg = 'red')
        messagebox.showwarning("Warning","Temperatura RF2 elevada")
        
    else:
        Temp_value_RF2.config(bg = "#00FF00")
        
    t_RF3.set(RF3)
    if RF3 > 22:
        Temp_value_RF3.config(bg = 'red')
        messagebox.showwarning("Warning","Temperatura RF3 elevada")
        
    else:
        Temp_value_RF3.config(bg = "#00FF00")
        
    t_RF4.set(RF4)
    if RF4 > 22:
        Temp_value_RF4.config(bg = 'red')
        messagebox.showwarning("Warning","Temperatura RF4 elevada")
        
    else:
        Temp_value_RF4.config(bg = "#00FF00")
        
    t_RF5.set(RF5)
    if RF5 > 22:
        Temp_value_RF5.config(bg = 'red')
        messagebox.showwarning("Warning","Temperatura RF5 elevada")
        
    else:
        Temp_value_RF5.config(bg = "#00FF00")
        
    t_SRVR1.set(SRVR1)
    if SRVR1 > 22:
        Temp_value_SRVR1.config(bg = 'red')
        messagebox.showwarning("Warning","Temperatura SRVR1 elevada")
        
    else:
        Temp_value_SRVR1.config(bg = "#00FF00")
        
    t_SRVR2.set(SRVR2)
    if SRVR2 > 22:
        Temp_value_SRVR2.config(bg = 'red')
        messagebox.showwarning("Warning","Temperatura SRVR2 elevada")
        
    else:
        Temp_value_SRVR2.config(bg = "#00FF00")

        
window = Tk()
window.title('Temperaturas')
window.geometry('550x60')

def update_values(window):
    chck_temp()
    insert()
    window.after(300000,update_values,window)

frame0 = Frame(window)    
frame0.pack(side=TOP, fill=X)
time_lbl = Label(frame0, text = "Latest Update: ")
time_lbl.pack(side = LEFT, padx = 5)
ult_act = StringVar()
Time_lbl = Label(frame0, textvariable = ult_act)
Time_lbl.pack(side = LEFT)

frame1 = Frame(window)
frame1.pack(side=TOP, fill=X)
Temp_lbl_RF1 = Label(frame1, text = "RF01:")
Temp_lbl_RF1.pack(side = LEFT, padx = 5, pady = 5)
t_RF1 = StringVar()
Temp_value_RF1 = Label(frame1, textvariable = t_RF1)
Temp_value_RF1.pack(side= LEFT, padx=5)

Temp_lbl_RF2 = Label(frame1, text = "RF02:")
Temp_lbl_RF2.pack(side = LEFT, padx = 5, pady = 5)
t_RF2 = StringVar()
Temp_value_RF2 = Label(frame1, textvariable = t_RF2)
Temp_value_RF2.pack(side= LEFT, padx=5)

Temp_lbl_RF3 = Label(frame1, text = "RF03:")
Temp_lbl_RF3.pack(side = LEFT, padx = 5, pady = 5)
t_RF3 = StringVar()
Temp_value_RF3 = Label(frame1, textvariable = t_RF3)
Temp_value_RF3.pack(side= LEFT, padx=5)

Temp_lbl_RF4 = Label(frame1, text = "RF04:")
Temp_lbl_RF4.pack(side = LEFT, padx = 5, pady = 5)
t_RF4 = StringVar()
Temp_value_RF4 = Label(frame1, textvariable = t_RF4)
Temp_value_RF4.pack(side= LEFT, padx=5)

Temp_lbl_RF5 = Label(frame1, text = "RF05:")
Temp_lbl_RF5.pack(side = LEFT, padx = 5, pady = 5)
t_RF5 = StringVar()
Temp_value_RF5 = Label(frame1, textvariable = t_RF5)
Temp_value_RF5.pack(side= LEFT, padx=5)

Temp_lbl_SRVR1 = Label(frame1, text = "SRVR1:")
Temp_lbl_SRVR1.pack(side = LEFT, padx = 5, pady = 5)
t_SRVR1 = StringVar()
Temp_value_SRVR1 = Label(frame1, textvariable = t_SRVR1)
Temp_value_SRVR1.pack(side= LEFT, padx=5)

Temp_lbl_SRVR2 = Label(frame1, text = "SRVR2:")
Temp_lbl_SRVR2.pack(side = LEFT, padx = 5, pady = 5)
t_SRVR2 = StringVar()
Temp_value_SRVR2 = Label(frame1, textvariable = t_SRVR2)
Temp_value_SRVR2.pack(side= LEFT, padx=5)

update_values(window)
window.mainloop()
```
