```python
from Crypto.Hash import SHA
```

```python
usuario = "mexsat"
contra = "mexsat"

h = SHA.new()
h.update(contra.encode())

contra_sha = h.hexdigest()
contra_sha
```

```python
from requests import Session
from bs4 import BeautifulSoup as bs

base_url = "http://192.168.169.3:81/"
login_data = {"uid_in":"mexsat","pid_in":"", "hex_in":contra_sha}

with Session() as s:
    
    response1 = s.post(base_url + ".html?get=${|mexsat|}", login_data)
    doc1 = bs(response1.text, 'html.parser')
    script1 = doc1.find_all("script")[0].text
    pag, user, sessionID = findall(".*\$(\w*)\|(\w*)\|(\w*)';", script1)[0]
    response_header = f".html?get=${pag}|{user}|{sessionID}"
    
    response2 = s.get(base_url + response_header)
    doc2 = bs(response2.text, 'html.parser')
    script2 = doc2.find_all("script")[0].text
    sessionID = findall('.*var sessionID = "([^\n\r]*)";', script2)[0]
    response_header = f".html?get=$AHU01|{user}|{sessionID}"
    
    response3 = s.get(base_url + response_header)
    doc3 = bs(response3.text, 'html.parser')
    script3 = doc3.find_all("script")[0].text
    sessionID = findall('.*var sessionID = "([^\n\r]*)";', script3)[0]
    response_header = f".html?get=$AHU02|{user}|{sessionID}"
```
