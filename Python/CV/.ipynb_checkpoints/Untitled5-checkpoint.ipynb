```python
from scipy.signal import chirp
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0, 0.5, 1001)
w = chirp(t, f0=6, f1=1, t1=10, method='linear')
plt.plot(t, w)
plt.title("Linear Chirp, f(0)=6, f(10)=1")
plt.xlabel('t (sec)')
plt.show()
```
