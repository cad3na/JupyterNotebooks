```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib.pyplot import subplots
from ipywidgets import interact, fixed
```

```python
def deteccion_bordes(imagen, v_inf, v_sup):
    img = cv2.imread(imagen, 0)
    edges = cv2.Canny(img, v_inf, v_sup)

    f, (ax1, ax2) = subplots(1, 2, sharey=True, figsize=(16,8))
    ax1.imshow(img, cmap = 'gray')
    ax1.set_title('Original Image')
    ax1.set_xticks([])
    ax1.set_yticks([])

    ax2.imshow(edges, cmap = 'gray')
    ax2.set_title('Edge Image')
    ax2.set_xticks([]);
```

```python
deteccion_bordes('prueba.jpg', 100, 200)
```

```python
interact(deteccion_bordes, imagen=fixed("prueba2.jpg"), v_inf = (0, 1000), v_sup = (0, 1000))
```
