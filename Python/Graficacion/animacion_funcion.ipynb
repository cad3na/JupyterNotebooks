```python
x, y = 0.5, 0.2
Δx, Δy = 0.8, 0.4

q = linspace(0,1.8, 100)

fig = figure(figsize=(8*1.8, 8))
ax = fig.gca()

datos = Circle((0, y+Δy/2), radius=0.05, lw=1.5, color="k")
ax.add_patch(datos)

funcion = Rectangle((x, y), Δx, Δy, lw=1.5, color="k", fc="white")
ax.add_patch(funcion)

ax.text(x + Δx/2, y + Δy/2, "funcion",
        horizontalalignment='center',
        verticalalignment='center',
        fontsize=50)

ax.quiver([0], [y+Δy/2], [x], [0],
          scale=1, scale_units="xy",
          angles="xy", color="k")
ax.quiver([x+Δx], [y+Δy/2], [0.5], [0],
          scale=1, scale_units="xy",
          angles="xy", color="k")

ax.axes.spines["top"].set_color("none")
ax.axes.spines["right"].set_color("none")
ax.axes.spines["bottom"].set_color("none")
ax.axes.spines["left"].set_color("none")

ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)

ax.set_xlim(0, 1.8)
ax.set_ylim(0, 0.8);

def init():
    datos.center = (0, y+Δy/2)
    funcion.set_xy([x,y])
    return datos

def animate(i):
    datos.center = (q[i], y+Δy/2)
    funcion.set_xy([x,y])
    if q[i] > 1:
        datos.set_facecolor("w")
    return datos

ani = animation.FuncAnimation(fig, animate, arange(1, len(q)), interval=30, init_func=init)
ani.save('./animacion-funcion.gif', writer='imagemagick');
#ani.save("./animacion_funcion.mp4")
```
