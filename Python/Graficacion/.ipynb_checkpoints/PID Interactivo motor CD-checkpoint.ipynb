```python
%matplotlib inline
from matplotlib.pyplot import figure, style
from control import tf, step_response
from numpy import linspace
style.use("ggplot")
```

```python
from ipywidgets import interact
```

```python
G = tf([1], [1,2,1])
t = linspace(0, 7, 1000)
```

```python
t,y = step_response(G, t)
```

```python
fig = figure(figsize=(15,8))
ax = fig.gca()
ax.plot(t,y)
```

```python
T = 3.2
L = 0.2
Kp = 1*(T/L)
Ti = 2*L
Td = L
```

```python
G1 = Kp*tf([Td*Ti, Ti,1], [Ti, 0])
```

```python
t,y = step_response((G1*G).feedback(), t)
fig = figure(figsize=(15,8))
ax = fig.gca()
ax.plot(t,y)
```

```python
def ejemplo(kp):
    t = linspace(0,1,100)
    t,y = step_response((kp*G).feedback(), t)
    fig = figure(figsize=(15,8))
    ax = fig.gca()
    ax.plot(t,y)
```

```python
interact(ejemplo, kp=5.0)
```
