```python
from matplotlib.pyplot import figure, style
style.use("ggplot")
%matplotlib inline
```

```python
from numpy import array, linspace
```

```python
def señal_PWM(DC, f, ts):
    from numpy import array
    T = 1/f
    T_on = T*DC
    sal = []
    for t in ts:
        if t==0:
            sal.append(0)
        else:
            if t%T < T_on:
                sal.append(1)
            else:
                sal.append(0)
    return array(sal)
```

```python
s1 = lambda t: señal_PWM(0.5, 10, t)
```

```python
ts = linspace(0, 2, 1000)
fig = figure(figsize=(20,5))
ax = fig.gca()
ax.plot(ts, s1(ts));
```
