```python
from openpyxl import load_workbook
wb = load_workbook(filename='catalogo-hsbc.xlsx', read_only=True)
ws = wb['Catálogo']

for row in ws.rows:
    for cell in row:
        print(cell.value)
```

```python
wb.get_sheet_names()
```

```python
filas = []
for row in ws.rows:
    celdas = []
    for celda in row:
        celdas.append(celda.value)
    filas.append(celdas)
```

```python
A01 = []
for fila in filas:
    if fila[0] == 'A01':
        A01.append(fila[2:-2])
```

```python
categorias = range(1, 20)
for i in categorias:
    print(str("%02d" % i))
```

```python
A = []
for categoria in categorias:
    Axx = []
    for fila in filas:
        if fila[0] == 'A' + str("%02d" % categoria):
            Axx.append(fila[2:-2])
        A.append(Axx)
```
