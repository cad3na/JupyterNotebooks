```python
from matplotlib.pyplot import figure, style
style.use("ggplot")
%matplotlib inline
```

```python
from numpy import sin, cos, pi, sqrt, radians, linspace
τ=2*pi
```

```python
f = 60
Vrms = 110
ω = τ*f
Vp = Vrms*sqrt(2)
ω, Vp
```

```python
V1 = lambda t: Vp*sin(ω*t)
V2 = lambda t: Vp*sin(ω*t - radians(120))
V3 = lambda t: Vp*sin(ω*t + radians(120))
```

```python
t = linspace(0, 0.1, 1000)
```

```python
fig = figure(figsize=(22, 4))
ax = fig.gca()
p1, = ax.plot(t, V1(t))
p2, = ax.plot(t, V2(t))
p3, = ax.plot(t, V3(t))
ax.legend([p1, p2, p3], [r"V_1", r"V_2", r"V_3"]);
```

```python
Va = lambda t: V1(t) - V2(t)
Vb = lambda t: V1(t) - V3(t)
```

```python
fig = figure(figsize=(22, 4))
ax = fig.gca()
p1, = ax.plot(t, Va(t))
p2, = ax.plot(t, Vb(t));
ax.legend([p1, p2], [r"V_A", r"V_B"]);
```

```python
max(Va(t))
```

```python
max(Va(t))/sqrt(2)
```

```python
fig = figure(figsize=(22, 4))
ax = fig.gca()
ax.plot(t, Va(t) - Vb(t))
```

```python
max(Va(t) - Vb(t))/sqrt(2)
```
