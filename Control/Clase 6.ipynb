```python
from control import tf, bode, nyquist, rlocus
figsize(11,8)
```

```python
sistema = tf([0, 0, 1], [1, 1, 0])
```

```python
mag, phase, omega = bode(sistema)
```

```python
real, imag, freq = nyquist(sistema)
```

```python
T1 = 1
T2 = 2
```

```python
sistema = tf([0, 0, 1], [0, T1, 1])*tf([0, 0, 1], [0, T2, 1])*tf([0, 0, 1], [0, 1, 0])
```

```python
r, k = rlocus(sistema)
ax = gca()
ax.set_xlim(-2.0, 1.0)
ax.set_ylim(-4.0, 4.0);
```

```python
real, imag, freq = nyquist(sistema)
ax = gca()
ax.set_xlim(-3.0, 0.5)
ax.set_ylim(-1.5, 1.5);
```

```python
real, imag, freq = nyquist(tf([3],[1])*sistema)
ax = gca()
ax.set_xlim(-3.0, 0.5)
ax.set_ylim(-1.5, 1.5);
```
