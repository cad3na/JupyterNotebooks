# Problem 1 a

$\{z\in\mathbb{C}\mid |z-3-2i|\leq 1\}$

If we define $z = x + y i$ then we would have:

$|x + y i - 3 - 2 i| \leq 1$

$|(x-3) + (y - 2) i| \leq 1$

$\sqrt{(x - 3)^2 + (y - 2)^2} \leq 1$

$(x-3)^2 + (y-2)^2 \leq 1$

So, in this example, we have a disk and its boundary in the definition of the set. The disk has center $(3, 2)$ and radius $1$.

```python
cir = Circle((3,2), 1, fc="b", ec="b")
fig = gcf()
fig.gca().add_artist(cir)
ax = gca()
ax.annotate("The disk", (3,2), (4,3), arrowprops=dict(arrowstyle="fancy", connectionstyle="arc3", fc="cyan", ec="cyan"))
ax.annotate("The boundary", (3,3), (4,3.5), arrowprops=dict(arrowstyle="fancy", connectionstyle="arc3", fc="cyan", ec="cyan"))
ax.annotate("This set includes the disk and the boundary", (0.5,0.5))
ax.set_xlim((0,6))
ax.set_ylim((0,4))
xlabel("Real axis")
ylabel("Imaginary axis")
grid(True)
```

# Problem 1 b

$\{z\in\mathbb{C}\mid \textrm{Im } z = 2\}$

If we define $z = x + y i$ then we would have:

$z = x + 2 i$

Where $x$ can have any value we want, and the imaginary part of $z$ is $2$ always. This describes a horizontal line.

```python
plot((0,6),(2,2))
fig = gcf()
ax = gca()
ax.annotate("The set", (3,2), (4,3), arrowprops=dict(arrowstyle="fancy", connectionstyle="arc3", fc="cyan", ec="cyan"))
ax.annotate("This set is the line", (0.5,0.5))
ax.set_xlim((0,6))
ax.set_ylim((0,4))
xlabel("Real axis")
ylabel("Imaginary axis")
grid(True)
```

# Problem 1 c

$\{z\in\mathbb{C}\setminus\{0\}\mid 0\lt \textrm{arg } z\lt \pi/6\}$

If $z = x + y i$ is a complex number, the $\arg{z}$ is the angle at which the number can be.

$\phi = \arg{z}$

Then $\phi$ can take any value between $0$ and $\frac{\pi}{6}$, and the modulus $|z|$ can be of any value.

Then the set will be all of the numbers between the boundaries $\arg{z} = 0$ and $\arg{z} = \frac{\pi}{6}$, not including the boundaries.

```python
plot1 = plot((0,6/cos(pi/6)),(0,2/sin(pi/6)))
fig = gcf()
ax = gca()
ax.fill_between((0,6/cos(pi/6)),(0,0),(0,2/sin(pi/6)))
ax.annotate("The set", (4,1), (2,2.5), arrowprops=dict(arrowstyle="fancy", connectionstyle="arc3", fc="cyan", ec="cyan"))
ax.annotate("0", (1,0), (0.5,1), arrowprops=dict(arrowstyle="fancy", connectionstyle="arc3", fc="cyan", ec="cyan"))
ax.annotate("pi/6", (5,3), (4,3), arrowprops=dict(arrowstyle="fancy", connectionstyle="arc3", fc="cyan", ec="cyan"))
ax.annotate("This set is all that region\n to the right without the boundaries", (1,3.5))
ax.set_xlim((0,6))
ax.set_ylim((0,4))
xlabel("Real axis")
ylabel("Imaginary axis")
grid(True)
```

# Problem 1 d

$\{z\in\mathbb{C}\mid |z-1|\lt|z|\}$

If $z = x + y i$ then:

$|x + y i - 1| \lt |z|$

$\sqrt{(x-1)^2 + y^2} \lt \sqrt{x^2 + y^2}$

$x^2 - 2 x + 1 + y^2 \lt x^2 + y^2$

$1 \lt 2x$

$x \gt \frac{1}{2}$

Therefore, the only restriction of the set is that the real part of the number, $x$, is greater than $\frac{1}{2}$. This is every number to the right of a vertical line in $\frac{1}{2}$.
