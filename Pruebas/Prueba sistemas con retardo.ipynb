```python
%matplotlib inline
from matplotlib.pyplot import plot, style, figure
style.use("ggplot")
```

```python
from control.delay import pade
from control import step, tf, rlocus
from numpy import linspace
```

```python
t = linspace(0, 2.5, 300)
G = tf([0, 1], [1, 10])
deltf = pade(1, n=30)
delay = tf(deltf[0], deltf[1])
ys, ts = step(10*G, t)
yd, td = step(10*delay*G, t)
```

```python
f = figure(figsize=(18, 5))
plot(ts, ys)
plot(td, yd);
```

```python
r, k = rlocus(delay*G);
```
