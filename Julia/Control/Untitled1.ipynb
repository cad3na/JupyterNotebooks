```julia
s = tf("s")
```

```julia
servo_hidraulico = 1/s
avion = (2*s + 0.1)/(s^2 + 0.1s + 4)
giroscopio = tf(1)
```

```julia
sistema = feedback(servo_hidraulico*avion, giroscopio)
```

```julia
stepplot(sistema, 500)
```
