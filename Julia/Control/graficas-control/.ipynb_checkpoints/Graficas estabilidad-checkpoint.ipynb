```julia
s = tf("s")
```

```julia
kp = 10
ki = 10
kd = 10

Control = kp + s*kd + ki/s
```

```julia
G = tf([1], [1,1,1])
ts = range(0, stop=10, length=100)
y = [exp(-0.5t+0.2)+0.05 for t=ts]
impulseplot(G, 10)
plot!(ts, y, line=([:dash], [:grey]))
xlabel!("")
ylabel!("")
title!("")
ylims!(-0.15,0.7)
plot!(frame_style=:origin)
savefig("est_exp.svg")
savefig("est_exp.pdf")
```

```julia
G = tf([2], [5,1,3])
ts = range(0, stop=20, length=100)
y = [-0.0205t + 0.48 for t=ts]
impulseplot(G, 20)
plot!(ts, y, line=([:dash], [:grey]))
xlabel!("")
ylabel!("")
title!("")
ylims!(-0.35,0.5)
plot!(frame_style=:origin)
savefig("est_asin.svg")
savefig("est_asin.pdf")
```
