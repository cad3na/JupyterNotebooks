```julia
using ControlSystems
using Plots
using Interact
using OrdinaryDiffEq

gr(size = (400,400), legend = false)
```

```julia
Ï‰s = exp10.(range(-1, stop=3, length=100))
s = tf("s")
```

```julia
bodeplot(1/s, Ï‰s, w=2)
xlabel!("")
ylabel!("")
title!("")
#savefig("bode-der.svg")
#savefig("bode-der.pdf")
```

```julia
bodeplot([tf(1), tf(10)], Ï‰s, w=2)
xlabel!("")
ylabel!("")
title!("")
```

```julia
bodeplot([s, 10*s], Ï‰s, w=2)
xlabel!("")
ylabel!("")
title!("")
```

```julia
bodeplot([1/s, 10/s], Ï‰s, w=2)
xlabel!("")
ylabel!("")
title!("")
```

```julia
bodeplot(1/(s+1), Ï‰s, w=2)
xlabel!("")
ylabel!("")
title!("")
```

```julia
bodeplot([10/(s*(s+1)), feedback(((s+2)/(s+1))*10/(s*(s+1)))], Ï‰s, w=2)
xlabel!("")
ylabel!("")
title!("")
```

```julia
bodeplot([1/(s+1), 1/((s+1)^2)], Ï‰s, w=2)
xlabel!("")
ylabel!("")
title!("")
```

```julia
bodeplot(100/((s^2+0.01*s+1)*(s+10)), Ï‰s, w=2)
xlabel!("")
ylabel!("")
title!("")
```

```julia
ui = @manipulate for Ï‰ in 0.1:0.01:5
    ts = 0.0:0.01:100.0
    u = [sin(Ï‰*t) for t in ts]
    lsimplot(100/((s^2+0.01*s+1)*(s+10)), u, ts, size = (1200,400))
end
```
