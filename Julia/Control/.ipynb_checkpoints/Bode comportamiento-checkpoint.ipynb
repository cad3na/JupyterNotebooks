```julia
using ControlSystems
using Plots
using Interact
using OrdinaryDiffEq
using LaTeXStrings

gr(size = (900,300))
```

```julia
ωs = exp10.(range(-1, stop=3, length=1000))
s = tf("s")
```

```julia
g(s) = 10/(s*(s+1))
```

```julia
G = g(s)
```

```julia
ui = @manipulate for ω in 0:0.01:5
    ts = 0.0:0.01:20.0
    u = [sin(ω*t) for t in ts]
    lsimplot(G, u, ts, frame_style=:origin, legend=:none)
    xlabel!("")
    ylabel!("")
    title!("")
end
```
