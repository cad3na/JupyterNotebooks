```julia
using Modia
using Modia.Electric
using ModiaMath
using Plots
```

```julia
@model LPfilter begin
  R=Resistor(R=100.0)
  C=Capacitor(C=2.5E-3)
  V=SineVoltage(V=10.0)
@equations begin
  connect(V.p, R.p)
  connect(R.n, C.p)
  connect(C.n, V.n)
  end
end;
```

```julia
result = simulate(LPfilter, 10);
```

```julia
Plots.plot(result["time"], result["C.v"])
Plots.plot!(result["time"], result["V.v"])
plot!(frame_style=:origin)
```
