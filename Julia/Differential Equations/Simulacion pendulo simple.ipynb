```julia
using DifferentialEquations
using Plots
```

```julia
function pend_simple(du, u, p, t)
    m, J, l = 0.1, 0.001, 0.1
    g = 9.81
    b = 0.002
    τ = 0
    du[1] = u[2]
    du[2] = (τ - b*u[2] - g*m*l*sin(u[1]))/(m*l^2 + J)
end
```

```julia
u0 = [1.0 ; 0.0]
tspan = (0.0, 10.0)
prob = ODEProblem(pend_simple, u0, tspan)
sol = solve(prob)
```

```julia
plot(sol, vars=(0, 1))
```
